For the maze shown in the following figure, the player can move one step each time in four directions with a cost: Up (5), Down (5), Left (1) and Right (1). The player cannot move into a block that it has previously visited. Find the optimal path for the player by using the A* search algorithm.


```python
graph1 = {
        'A': [('B', 1), ('F', 5)],
        'B': [('A', 1), ('C', 1)],
        'C': [('D', 1), ('B', 1)],
        'D': [('C', 1), ('E', 1)],
        'E': [('D', 1), ('G', 5)],
        'F': [('A', 5), ('H', 5)],
        'G': [('E', 5), ('J', 5)],
        'H': [('F', 5), ('K', 5)],
        'I': [('L', 5)],
        'J': [('G', 5), ('N', 5)],
        'K': [('H', 5), ('O', 5)],
        'L': [('I', 5), ('M', 1), ('Q', 5)],
        'M': [('L', 1), ('N', 1), ('R', 5)],
        'N': [('J', 5), ('M', 1), ('S', 5)],
        'O': [('K', 5), ('P', 1)],
        'P': [('O', 1), ('Q', 1)],
        'Q': [('L', 5), ('P', 1), ('R', 1)],
        'R': [('M', 5), ('Q', 1), ('S', 1)],
        'S': [('N', 5), ('R', 1)]
    }

path = a_star_search(graph1, start='I', end='A')
print(f"Path found: {path}")
```
